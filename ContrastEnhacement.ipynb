```python
import skimage.io as io
import numpy as np
from commonfunctions import *
```

```python
def globalHistogramEqualization(imgDir):
    imgCopy =np.copy(io.imread(imgDir))
    img=np.copy(io.imread(imgDir))
    size = img.shape[0]*img.shape[1]
    imgHist = showHist(img)
    cummulative = np.cumsum(imgHist[0])
    probCummulative = np.round(255*(cummulative/size))
    imgnotEqualized = np.copy(img)
    for i in range(imgHist[1].shape[0]):
        img[(np.where(imgnotEqualized[:] == imgHist[1][i]))] = probCummulative[i]
    showHist(img)
    show_images([imgCopy, img])

globalHistogramEqualization("DataSets/LB1.png")
```
